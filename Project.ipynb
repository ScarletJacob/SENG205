```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
import pandas as pd
redWine= pd.read_csv('red_wine.csv')
df = pd.DataFrame(redWine)
density_column=df.density.head(10)
pH_column=df.pH.head(10)
alcohol_column=df.alcohol.head(10)

plt.figure(figsize=(5, 6))  # Set the figure size

#Plotting density against pH
plt.plot(density_column,pH_column, label='pH',linestyle='-', marker='o')

#Plotting density against alcohol

plt.plot(density_column,alcohol_column , label='Alcohol',linestyle='-', marker='*')

plt.title('Line Plot for Two Columns')
plt.xlabel('Density')
plt.ylabel('Alcohol and pH')
plt.legend() 
plt.show()
```

```python
plt.figure(figsize=(10, 6))  # Set the figure size

plt.bar(density_column - 0.2,pH_column , width=0.2, label='pH', alpha=0.7)

plt.bar(density_column + 0.2,alcohol_column , width=0.2, label='Alcohol', alpha=0.7)

plt.title('Bar Chart for Density against pH and alcohol')
plt.xlabel('Density')
plt.ylabel('pH and Alcohol')
plt.legend()  # Show legend

# Display the bar chart
plt.show()
```

# <center> REPORTS ON PLOTS AND GRAPH OBTAINED FROM DATA </center>

# <center> Report On Lineplot </center>

It was observed that even though both pH and alcohol from data were both plotted against density, there was no correlation between *__pH__* and *__alcohol__*. This can be further seen from the plots- two distinct plots  attained. It can also be said that both pH and alcohol plottings were quite random, seemingly scatter-looking

## <center> Report On Bar Chart </center>

On the bar chart, two different bars can be seen; both of different size and color. Aside from that it can be seen that both appear to have apparent lighter shades of their respective colors behind them. Those shades are accounting for the other numbers of *__pH__* and *__Alcohol__* as well.
